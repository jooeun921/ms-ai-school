# 브라우저 제어
    - `get(URL)` : 페이지로 이동하기기
    - `back()` : 뒤로가기
    - `forward()` : 앞으로가기
    - `refresh()` : 새로고침
    - `maximize_window()` : 브라우저크기 최대화
    - `minimize_window()` : 브라우저크기 최소화
    - `close()` : 웹드라이버로 생성한 현재 창 종료
    - `quit()` : 웹드라이버로 생성한 모든 창 종료

```python
url = 'https://finance.naver.com/news/mainnews.naver'

driver = Chrome()
driver.get(url)
```

```python
from selenium.webdriver.common.by import By

articles = driver.find_elements(By.CSS_SELECTOR, ".block1")

article_list = []
for article in articles:
    subject = article.find_element(By.CSS_SELECTOR, ".articleSubject").text
    summary = article.find_element(By.CSS_SELECTOR, ".articleSummary").text
    press = article.find_element(By.CSS_SELECTOR, ".press").text
    wdate = article.find_element(By.CSS_SELECTOR, ".wdate").text
    link = article.find_element(By.CSS_SELECTOR, ".articleSubject>a").get_attribute("href")
    article_list.append([subject, summary, press, wdate, link])
    print(article_list)
```

```python
import pandas as pd

df = pd.DataFrame(article_list, columns=['subject', 'summary', 'press', 'wdate', 'link'])
df.to_csv("네이버증권뉴스.csv", encoding='utf-8', index=False)
df.to_excel("네이버증권뉴스.xlsx", index=False)

```

```python
import datetime

# 날짜 입력 받아오기. 공백이면 오늘 날짜로 받고, 입력한 string 값이 있으면 슬라이싱으로 자른 뒤에 사용
# https://finance.naver.com/news/mainnews.naver?date=2025-01-29&page=3
date=input("검색할날짜(yyyymmdd) : ")
if date=="": date=datetime.datetime.now().strftime('%Y-%m-%d')
else: date=f'{date[:4]}-{date[4:6]}-{date[6:]}'

from selenium.webdriver import Chrome
import time

driver=Chrome()

article_list = []
page = 1

while True:
    #페이지 이동. 주소 살펴보게 되면, 페이지가 이동할 때마다 &page=1, &page=2... 식으로 변함.
    url="https://finance.naver.com/news/mainnews.naver?date="+date+"&page="+str(page)
    driver.get(url)

    time.sleep(2)

    articles = driver.find_elements(By.CSS_SELECTOR, ".block1")

    for article in articles:
        subject = article.find_element(By.CSS_SELECTOR, ".articleSubject").text
        summary = article.find_element(By.CSS_SELECTOR, ".articleSummary").text
        press = article.find_element(By.CSS_SELECTOR, ".press").text
        wdate = article.find_element(By.CSS_SELECTOR, ".wdate").text
        link = article.find_element(By.CSS_SELECTOR, ".articleSubject>a").get_attribute("href")
        article_list.append([subject, summary, press, wdate, link])

    # 반복을 빠져나오는 조건. 맨끝 버튼이 없어졌을 때가 마지막 페이지임
    try:
        driver.find_element(By.CSS_SELECTOR, ".pgRR")
        page += 1
    except:
        break
```

```python
df = pd.DataFrame(article_list, columns=['subject', 'summary', 'press', 'wdate', 'link'])
df.to_csv(f"네이버증권뉴스_{date}.csv", encoding='utf-8', index=False)
df.to_excel(f"네이버증권뉴스_{date}.xlsx", index=False)
```

```python
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.common.keys import Keys
import pandas as pd
import time
import openpyxl
 
url = "https://search.shopping.naver.com/search/all?query=노트북"
driver = webdriver.Chrome()
driver.get(url)
driver.maximize_window()
 
 
# 무한스크롤
# 페이지의 스크롤 높이 받아오기
last_height = driver.execute_script("return document.body.scrollHeight")
while True:
    # 스크롤 끝까지 내리기
    driver.execute_script("window.scrollTo(0,document.body.scrollHeight)")
 
    # 현재 스크롤 위치 받아오기
    current_height = driver.execute_script("return document.body.scrollHeight")
 
    # 마지막스크롤 = 현재스크롤이면 반복 벗어나기  
    if last_height == current_height: break
 
    # 현재 스크롤을 마지막 스크롤로 업데이트 하기
    last_height = current_height
 
    time.sleep(1)
 
```

```python
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.common.exceptions import NoSuchElementException, StaleElementReferenceException
import time

def load_all_comments(driver, url):
    driver.get(url) # 지정된 URL로 브라우저를 열어 이동
    time.sleep(2) # 페이지 로드 대기

    while True:
        try:
            # "댓글 더보기" 버튼 찾기
            more_comments_button = driver.find_element(By.CSS_SELECTOR, '.u_cbox_in_view_comment')
            more_comments_button.click() # "댓글 더보기" 버튼 클릭
            time.sleep(2) # 댓글 로드 대기
        except (NoSuchElementException, StaleElementReferenceException):
            # "댓글 더보기" 버튼이 더 이상 없거나 요소 참조가 유효하지 않을 경우 반복문 탈출
            break

# 사용 예시
url = "https://n.news.naver.com/mnews/hotissue/article/028/0002728835?type=series&cid=2001999" # 네이버 뉴스 기사 URL
driver = webdriver.Chrome() # Chrome 웹드라이버 객체 생성

load_all_comments(driver, url) # 댓글을 모두 로드하는 함수 호출
```
