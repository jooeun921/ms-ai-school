```python
# 라이브러리 불러오기
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

# k-means clustering

### 데이터 준비
* 테스트 데이터 생성
    * make_blobs()

```python
# 3개의 군집을 가지는 테스트 데이터 생성
from sklearn.datasets import make_blobs
features, targets = make_blobs(n_samples=200, 
                    n_features=2,
                    centers=3,
                    cluster_std=0.8,
                    random_state=0)

print(features.shape, targets.shape)
```

```python
# 생성된 테스트데이터로 DataFrame 만들기
cluster_df = pd.DataFrame(features, columns=['feature1','feature2'])
cluster_df['target'] = targets
cluster_df.head(3)
```

```python
# target 데이터의 빈도수
cluster_df['target'].value_counts()
```

```python
# 데이터 시각화
sns.scatterplot(data=cluster_df, x='feature1', y='feature2', 
                hue='target', palette='muted');
plt.title('Original Data');
```

### 군집화

```python
# 클러스터링 객체 생성
from sklearn.cluster import KMeans
km=KMeans(n_clusters=3)

# 학습 및 클러스터링 결과 예측
# km.fit(features)
# kmeans_cluster= km.predict(features)

kmeans_cluster=km.fit_predict(features)
print(kmeans_cluster)
```

```python
# 데이터프레임에 군집과 결과인 kmeans_cluster 컬럼 추가
cluster_df['kmeans_cluster'] = kmeans_cluster
cluster_df.head(3)
```

```python
# 클러스터의 중심
km.cluster_centers_
```

```python
# 군집화 결과 시각화
sns.scatterplot(data=cluster_df, 
x='feature1', 
y='feature2', 
hue='kmeans_cluster', 
palette='muted')

# 개별 군집의 중심 좌표
sns.scatterplot(x=km.cluster_centers_[:,0], 
y=km.cluster_centers_[:,1], 
color='magenta', marker='X')
plt.title('k-means Clustering')
plt.show()
```

### elbow 방법

```python
# inertia_: 각 데이터 포인트와 그 데이터가 속한 군집의 중심점 사이의 거리 제곱합 
inertia =[]
for n in range(2,7):
    km =KMeans(n_clusters=n)
    km.fit(features)
    print(km.inertia_)
    inertia.append(km.inertia_)
```

```python
# inertia 시각화
plt.plot(range(2,7), inertia, marker='o')
plt.xticks(range(2,7))
plt.xlabel('k')
plt.ylabel('inertia')
plt.show()
```

### 실루엣 분석

```python
# 개별 실루엣 계수 (군집 평가 지표)
from sklearn.metrics import silhouette_samples

cluster_df['silhouette'] =silhouette_samples(features, kmeans_cluster)
cluster_df.head()
```

```python
# 다른 군집에 클러스터가 할당된 데이터가 있는지 확인
cond = cluster_df['silhouette'] < 0
cluster_df.loc[cond]
```

```python
# 평균 실루엣 계수
from sklearn.metrics import silhouette_score
silhouette_score(features, kmeans_cluster)
```

```python
# 최적의 실루엣 찾기
import silhouette_analysis as s
for k in range(2,7):
    s.silhouette_plot(features, k)
```

# DBSCAN

### 데이터 준비

```python
from sklearn.datasets import make_circles

X, y = make_circles(n_samples=1000,
                    shuffle=True,
                    noise=0.05,
                    random_state=0,
                    factor=0.5)

df = pd.DataFrame(X, columns=['feature1', 'feature2'])
df['target'] = y

df.head()
```

```python
sns.scatterplot(data=df, x='feature1', y='feature2');
```

### k-means clustering

```python
from sklearn.cluster import KMeans

# 클래스로 객체 생성
km = KMeans(n_clusters=2, random_state=0)

# 클러스터링
df['kmeans'] = km.fit_predict(X)
df.head()
```

```python
sns.scatterplot(data=df, x='feature1', y='feature2', hue='kmeans');
sns.scatterplot(x=km.cluster_centers_[:,0], 
                y=km.cluster_centers_[:,1],
                marker='X', color='k');
```

### DBSCAN

```python
from sklearn.datasets import make_circles
X, y =make_circles(n_samples=1000, shuffle=True, noise=0.05, random_state=0, factor=0.5)
df=pd.DataFrame(X, columns=['feature1', 'feature2'])
df['target'] =y
df.head()
```

```python
sns.scatterplot(data=df, x='feature1', y='feature2', hue='target', palette='muted')
plt.show()
```

```python
from sklearn.cluster import KMeans
km =KMeans(n_clusters=2)
df['kmeans'] =km.fit_predict(X)
df.head()
```

```python
sns.scatterplot(df, x='feature1', y='feature2', hue='kmeans', palette='muted')
sns.scatterplot(x=km.cluster_centers_[:,0], y=km.cluster_centers_[:,1],  marker='D', color='magenta')
plt.title('k-means clustering')
plt.show()
```

```python
from sklearn.cluster import DBSCAN
dbscan=DBSCAN(eps=0.2,
min_samples=10)
df['dbscan'] =dbscan.fit_predict(X)
df.head()
```

```python
sns.scatterplot(df, x='feature1', y='feature2', hue='dbscan')
plt.title('DBSCAN')
plt.show()
```
