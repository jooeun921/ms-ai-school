```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
 
plt.style.use('ggplot')
pd.Series([1,2,3]).plot(figsize=(3,2))
```

```python
from sklearn.datasets import load_iris
 
# 붓꽃 데이터 로딩
iris = load_iris()

iris.keys() 
```

```python
# 독립변수
feature = iris.data
feature_name = iris.feature_names

print(feature_name)
print(feature)
```

```python
# 종속변수
target = iris.target
target_name = iris.target_names

print(target_name)
print(target)
```

```python
print(iris.DESCR)
```

```python
# 데이터프레임 만들기
df_iris = pd.DataFrame(feature, columns=feature_name)
df_iris['species'] = target
df_iris.head()
```

```python
df_iris.info()
```

```python
df_iris['species'].value_counts()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

sns.pairplot(data=df_iris, hue='species', palette='muted')
plt.show()
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(feature, target, test_size=0.2, random_state=42)
print(X_train.shape, X_test.shape, y_train.shape, y_test.shape)
```

```python
print(pd.Series(y_train).value_counts())
print(pd.Series(y_test).value_counts())
```

```python
X_train, X_test, y_train, y_test = train_test_split(feature, target, test_size=0.2, random_state=42, stratify=target)
print(X_train.shape, X_test.shape, y_train.shape, y_test.shape)
```

```python
print(pd.Series(y_train).value_counts())
print(pd.Series(y_test).value_counts())
```

```python
from sklearn.tree import DecisionTreeClassifier

model = DecisionTreeClassifier(random_state=42)

model.fit(X_train, y_train)
```

```python
pred = model.predict(X_test)
print("예측 label >>", pred)
print("실제 label >>", y_test)
```

```python
from sklearn.metrics import accuracy_score

accuracy = accuracy_score(y_test, pred)
print(f"예측 정확도 >> {accuracy:.2f}")
```

```python
from sklearn.metrics import classification_report

print(classification_report(y_test, pred))
```

```python
# KFold 객체 생성. 5개로 분할, 반복함. -> 이렇게 분할된 데이터를 활용해서 각각마다 모델을 훈련시켜서 가장 좋은 성능을 가진 애를 뽑을 것임.
from sklearn.model_selection import KFold

kf = KFold(n_splits=5, shuffle=True, random_state=42)
```

```python
# k개의 fold로 분할하여 k번 반복. 여러번 학습을 통해 정확도를 높이기 위한 방법.

for tr_idx, val_idx in kf.split(X_train):
    print(f'tr_idx >> {tr_idx}')
    print(f'val_idx >> {val_idx}')
```

```python
for tr_idx, val_idx in kf.split(X_train):
    X_tr, y_tr = X_train[tr_idx], y_train[tr_idx]
    X_val, y_val = X_train[val_idx], y_train[val_idx]
    print(X_tr.shape, X_val.shape, y_tr.shape, y_val.shape)
```

```python
for train_index, test_index in kf.split(feature):
    X_train, y_train = feature[train_index], target[train_index] # 학습용 데이터
    X_test, y_test = feature[test_index], target[test_index] # 검증용 데이터

    model.fit(X_train, y_train)
    pred = model.predict(X_test)
    accuracy = accuracy_score(y_test, pred)
    print(f'accuracy >> {accuracy}')
```

```python
cv_scores = []
for tr_idx, val_idx in kf.split(X_train):
    X_tr, y_tr = X_train[tr_idx], y_train[tr_idx] # 학습용 데이터
    X_val, y_val = X_train[val_idx], y_train[val_idx] # 검증용 데이터
    
    model.fit(X_tr, y_tr)
    pred = model.predict(X_val)
    accuracy = accuracy_score(y_val, pred)
    cv_scores.append(accuracy)

print(f'cv_scores: {np.round(cv_scores, 4)}')
print(f'cv_scores_mean: {np.mean(cv_scores):.4f}')
```

```python
cv_scores = []
for tr_idx, val_idx in kf.split(X_train):
    X_tr, y_tr = X_train[tr_idx], y_train[tr_idx] # 학습용 데이터
    X_val, y_val = X_train[val_idx], y_train[val_idx] # 검증용 데이터
    print(pd.Series(y_val).value_counts())
    
    model.fit(X_tr, y_tr)
    pred = model.predict(X_val)
    accuracy = accuracy_score(y_val, pred)
    cv_scores.append(accuracy)

print(f'cv_scores: {np.round(cv_scores, 4)}')
print(f'cv_scores_mean: {np.mean(cv_scores):.4f}')
```

```python
from sklearn.model_selection import StratifiedKFold

skf = StratifiedKFold(n_splits=5, shuffle=True, random_state=42)
```

```python
cv_scores = []
for tr_idx, val_idx in skf.split(X_train, y_train):
    X_tr, y_tr = X_train[tr_idx], y_train[tr_idx] # 학습용 데이터
    X_val, y_val = X_train[val_idx], y_train[val_idx] # 검증용 데이터
    print(pd.Series(y_val).value_counts())

    model.fit(X_tr, y_tr)
    pred = model.predict(X_val)
    accuracy = accuracy_score(y_val, pred)
    cv_scores.append(accuracy)

print(f'cv_scores: {np.round(cv_scores, 4)}')
print(f'cv_scores_mean: {np.mean(cv_scores):.4f}')
```

```python
# 교차검증 수행
from sklearn.model_selection import cross_val_score

cv_scores=cross_val_score(model, X=X_train, y=y_train, scoring='accuracy', cv=skf)
print(f'cv_scores: {np.around(cv_scores,4)}')
print(f'cv_scores_mean: {cv_scores.mean():.4f}')
```

```python
# 훈련데이터 전체로 훈련, 테스트데이터로 최종 평가

model.fit(X_train, y_train)
y_pred = model.predict(X_test)
accuracy = accuracy_score(y_test, y_pred)
print(f'test accuracy : {accuracy:.4f}')
```

```python
# 트리 시각화
from sklearn.tree import plot_tree

plt.figure(figsize=(20,10))
plot_tree(model, feature_names=feature_name, class_names=target_name, filled=True)
plt.tight_layout()
plt.show()
```

```python
# 특성의 중요도

print(f'특성의 중요도 >>> {model.feature_importances_}')
plt.figure(figsize=(4,3))
sns.barplot(y=feature_name, x=model.feature_importances_)
plt.show()
```

```python
# 하이퍼파라미터 준비
parameters ={'max_depth':[3,4,5], 'min_samples_split':[2,4]}

# GridSearchCV객체 생성
from sklearn.model_selection import GridSearchCV

gscv=GridSearchCV(model, param_grid=parameters, cv=kf, refit=True)

# 하이퍼파라미터를 순차적으로 변경하면서 학습/평가 수행
gscv.fit(X_train, y_train)
```

```python
# 그리드서치 결과 확인
display(pd.DataFrame(gscv.cv_results_))
print(f'최적의 파라미터 >>> {gscv.best_params_}')
print(f'최고 정확도 >>> {gscv.best_score_:.4f}')
```

```python
# 최종 모델
best_model=gscv.best_estimator_

# 예측
pred =best_model.predict(X_test)

# 평가
from sklearn.metrics import accuracy_score

accuracy =accuracy_score(y_test, pred)
print(f'테스트세트 정확도 >> {accuracy:.4f}')
```

```python
from sklearn.ensemble import RandomForestClassifier

model2 = RandomForestClassifier(random_state=42)
gscv2=GridSearchCV(model2, param_grid=parameters, cv=kf, refit=True)

gscv2.fit(X_train, y_train)
```

```python
# 그리드서치 결과 확인
display(pd.DataFrame(gscv2.cv_results_))
print(f'최적의 파라미터 >>> {gscv2.best_params_}')
print(f'최고 정확도 >>> {gscv2.best_score_:.4f}')
```

```python
# 최종 모델
best_model=gscv2.best_estimator_

# 예측
pred =best_model.predict(X_test)

# 평가
from sklearn.metrics import accuracy_score

accuracy =accuracy_score(y_test, pred)
print(f'테스트세트 정확도 >> {accuracy:.4f}')
```
